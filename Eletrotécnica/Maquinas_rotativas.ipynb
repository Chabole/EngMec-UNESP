```python
import numpy as np
import matplotlib.pyplot as plt

def coss(x):
    return np.cos(np.radians(x))

wt = np.linspace(1, 360)
r2 = 2**(1/2)

fig, ax = plt.subplots()
ax.set(title='Campo girante em maquinas el√©tricas', 
       xlabel=r'$h_x$', ylabel=r'$h_y$')

I = [(5, 5, 5), (5, 5, 0), (5, 10, 0)]

for i in I:
    
    #Correntes de fase
    Ia, Ib, Ic = i

    ia = r2*Ia*coss(wt)
    ib = r2*Ib*coss(wt-120)
    ic = r2*Ic*coss(wt+120)

    ha, hb, hc = 0.2*ia, 0.2*ib, 0.2*ic 

    hx =  ha + hb*(-1)*coss(60) + hc*(-1)*coss(60)
    hy =  0 + hb*(1)*coss(30) + hc*(-1)*coss(30)

    ax.plot(hx, hy, label=f'Ia={Ia}, Ib={Ib}, Ic={Ic}')
ax.grid(linestyle='dotted')
ax.legend()
```
